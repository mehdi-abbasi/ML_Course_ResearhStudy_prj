```python
import io
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
df
```

```python
df = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
print(df.isnull().sum().sum())
df.replace([np.inf, - np.inf], np.nan, inplace = True)
print(df.isnull().sum().sum())
df = df.dropna() 
df = df.reset_index(drop=True)
X = df._get_numeric_data().values
Y = df[' Label']
print(df.isnull().sum().sum())
print('List of labels:',list(set(Y)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,  test_size=0.3,random_state=1)
```

```python
mlp1 = MLPClassifier(random_state=5,hidden_layer_sizes=(100,))
mlp1.fit(X_train, y_train)
```

```python
mlp1.score(X_test, y_test)
```

```python
mlp2 = MLPClassifier(random_state=5,hidden_layer_sizes=(100,100))
mlp2.fit(X_train, y_train)
```

```python
mlp2.score(X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
```

```python
rf.score(X_test, y_test)
```

```python
mlp1_pred = mlp1.predict(X_test)
mlp2_pred = mlp2.predict(X_test)
rf_pred = rf.predict(X_test)
print('confusion_matrix_MLP1')
print(confusion_matrix(mlp1_pred,y_pred))
print('-----------------------------')
print('confusion_matrix_MLP2')
print(confusion_matrix(mlp2_pred,y_pred))
print('-----------------------------')
print('confusion_matrix_RF')
print(confusion_matrix(rf_pred,y_pred))
```
